# **BASE DE TREINO E TESTE** #

```python
from sklearn.model_selection import train_test_split
```

```python
#test_size: Tamanho em porcentagem dos dados de teste;

#random_state = 0: Garante que os resultado serão os mesmos toda vez que o código for executado;

#X = previsores

#Y = alvo

x_treino, x_teste, y_treino, y_teste = train_test_split(previsores, alvo, test_size=0.3, random_state=5, stratify= alvo)
```

```python
x_treino.shape
```

```python
x_teste.shape
```

```python
y_treino.shape
```

```python
y_teste.shape
```

# **NAIVE BAYES**

### Treinamento do algoritmo

```python
from sklearn.naive_bayes import GaussianNB
```

```python
naive = GaussianNB()
naive.fit(x_treino, y_treino)
```

```python
#Gerando as previsões do algoritmo
previsoes_naive = naive.predict(x_teste)
previsoes_naive
```

```python
#Verificando as respostas corretas (gabarito)
y_teste
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix
```

```python
confusion_matrix(y_teste, previsoes_naive)
```

```python
accuracy_score(y_teste, previsoes_naive)
```

```python
print("Acurácia: %.2f%%" % (accuracy_score(y_teste, previsoes_naive) * 100.0))
```

# **MÁQUINAS DE VETORES DE SUPORTE (SVM)**

```python
from sklearn.svm import SVC
```

```python
svm = SVC(kernel='rbf', random_state=1, C = 50)
```

```python
#Treinar o modelo SVC
svm.fit(x_treino, y_treino)
```

```python
#Testando o Modelo
previsoes_svm = svm.predict(x_teste)
```

```python
previsoes_svm
```

```python
y_teste
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix
```

```python
print("Acurácia: %.2f%%" % (accuracy_score(y_teste, previsoes_svm) * 100.0))
```

# Aprendizagem Baseada em Distâncias (KNN) 

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
#Aplicando o Modelo
knn = KNeighborsClassifier(n_neighbors=5, metric='euclidean')
```

```python
#Treinando o Modelo
knn.fit(x_treino, y_treino)
```

```python
#Aplicar Teste
previsoes_knn = knn.predict(x_teste)
```

```python
previsoes_knn
```

```python
y_teste
```

```python
print("Acurácia: %.2f%%" % (accuracy_score(y_teste, previsoes_knn) * 100.0))
```
